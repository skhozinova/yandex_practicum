```python
#объединим таблицы с полученными ранее расчётами в одну
call_message = number_of_minutes_by_month.join(number_of_messages_by_month)
call_message_internet = call_message.join(number_of_mb_by_month)

#общую таблицу объеиним с информацией о пользователях, чтобы узнать их тариф
total_user = call_message_internet.merge(data_users, on = 'user_id')
total_user.head(10)
```

```python
total.info()
```

```python
#проверка на дубликаты 
total.duplicated().sum()
```

```python
total['price_for_megabyte'] = total['rub_per_gb'] / 1024
total.sample(n=5)
```

```python
total.head(5)
```

```python
#вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; 
#остаток умножьте на значение из тарифного плана; 
#прибавьте абонентскую плату, соответствующую тарифному плану
def profit_of_megaline(row):
    internet = row['internet']
    tariff = row['tariff_name']
    calls = row['calls']
    messages = row['messages']
    mb_per_month_included = row ['mb_per_month_included']
    minutes_included = row['minutes_included']
    messages_included = row['messages_included']
    price_for_megabyte = row ['price_for_megabyte']
    rub_per_minute = row['rub_per_minute']
    rub_per_message = row['rub_per_message']
    rub_monthly_fee = row['rub_monthly_fee']
    
    
    
    total_internet = 0
    total_messages = 0 
    total_calls = 0
    
    if internet > mb_per_month_included:
        total_internet = (internet - mb_per_month_included) * price_for_megabyte
    if calls > minutes_included:
        total_calls = (calls - minutes_included) * rub_per_minute
    if messages > messages_included:
        total_messages = (messages - messages_included) * rub_per_message
        
    return total_internet + total_calls + total_messages + rub_monthly_fee
    
```

```python
total.sample(5)
```

```python
profit_smart = total.query('tariff_name == "smart" and profit > 0')
profit_ultra = total.query('tariff_name == "ultra" and profit > 0')
profit_smart_sum = profit_smart['profit'].sum()
profit_ultra_sum = profit_ultra['profit'].sum()
if profit_smart_sum > profit_ultra_sum:
    print('Тариф Smart приносит больше прибыли на', profit_smart_sum - profit_ultra_sum)
else: print('Тариф Ultra приносит больше прибыли на', profit_ultra_sum - profit_smart_sum)
```

## Проверка гипотез
- средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;
- средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;
Пороговое значение alpha задайте самостоятельно.

Поясните:
- как вы формулировали нулевую и альтернативную гипотезы;
- какой критерий использовали для проверки гипотез и почему.

### Проверить гипотезу, что средняя выручка пользователей тарифов «Ультра» и «Смарт» различается.

Нулевая гипотеза: Средняя выручка пользователей тарифов Смарт и Ультра не различается

Альтернативная гипотеза: Средняя выручка пользователей тарифов Смарт и Ультра различается

```python
alpha = 0.05

sample_1 = profit_smart['profit'] # sample_1 - выборка из первой генеральной совокупности
sample_2 = profit_ultra['profit'] # sample_2 - выборка из второй генеральной совокупности

results = st.ttest_ind(sample_1, sample_2)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Проверить гипотезу, что средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

Нулевая гипотеза: средняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов.

Альтернативная гипотеза: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

```python
moscow_profit = total.query('city == "Москва" and profit > 0')
other_city_profit = total.query('city != "Москва" and profit > 0')
other_city_profit.sample(10)
```

```python
alpha = 0.05

sample_1 = moscow_profit['profit'] # sample_1 - выборка из первой генеральной совокупности
sample_2 = other_city_profit['profit'] # sample_2 - выборка из второй генеральной совокупности

results = st.ttest_ind(sample_1, sample_2)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**

Проверка двух гипотез показала, что средняя выручка пользователей тарифов «Ультра» и «Смарт» различается и средняя выручка пользователей с Москвы от пользователей из других регионов не отличается.
