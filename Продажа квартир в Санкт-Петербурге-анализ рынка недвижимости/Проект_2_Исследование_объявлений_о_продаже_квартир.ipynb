## Предобработка данных

### Исследование и заполнение пропусков в данных

```python
#проверим количество NAN в столбцах
data.isna().sum()
```

```python
#заменим NAN в ceiling_height
data['ceiling_height']=data['ceiling_height'].fillna(data['ceiling_height'].median())
```

```python
data[data['floors_total'].isnull()]
```

```python
#удалим NAN в floors_total
data.dropna(subset=['floors_total'], inplace=True)
```

```python
#заменим NAN на медианную площадь для соответствующих групп комнат
data['living_area'] = data.groupby('rooms')['living_area'].transform(lambda x: x.fillna(x.median()))
```

```python
#заменим NAN в is_apartment
data['is_apartment'] = data['is_apartment'].fillna(False)
```

```python
data[data['kitchen_area'].isnull()]
```

```python
#заменим NAN в kitchen_area на медианную площадь для соответствующих групп комнат
data['kitchen_area'] = data.groupby('rooms')['kitchen_area'].transform(lambda x: x.fillna(x.median()))
#у оставшихся квартир с количеством комнат 0, заменим NAN на 0
data['kitchen_area']=data['kitchen_area'].fillna(0)
```

```python
#заменим NAN в balcony на 0, т.к если не указано число балконов, скорее всего их нет
data['balcony']=data['balcony'].fillna(0)
```

```python
#удалим строки с пропусками в столбце locality_name, их не так много
data.dropna(subset=['locality_name'], inplace=True)
```

```python
# заменим пропуски числа парков, если стоит пропуск, скорее всего парков в радиусе 3км нет
data['parks_around3000']= data['parks_around3000'].fillna(0)
```

```python
# тоже самое касается числа водоёмов 
data['ponds_around3000']= data['ponds_around3000'].fillna(0)
```

   Вывод
   
   Данные о расстоянии до центра, аэропорта, парков и водоёмов не будем изменять, все записи понадобятся в дальнейшем исследовании.

### Изменение типов данных

```python
data.info()
```

```python
#изменим тип данных столбца first_day_exposition со строчного на datetime и уберем время т.к. оно везде одинаковое и не несёт никакой информации
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d')
```

```python
#количество балконов, дней размещения, комнат, парков, водоёмов, этажей измеряется в целых значения, так почему бы не перевести их в int

 
data['rooms'] = data['rooms'].astype('int')
data['floor'] = data['floor'].astype('int')
data['floors_total'] = data['floors_total'].astype('int')
data['parks_around3000'] = data['parks_around3000'].astype('int')
data['ponds_around3000'] = data['ponds_around3000'].astype('int')
data.info()
```

## Рассчёты и добавление результатов в таблицу

### Посчитайте и добавьте в таблицу:
#### Цену квадратного метра.

```python
#найдем цену квадратного метра и округлим до десятых
data['price_per_square_meter'] = round(data['last_price'] / data['total_area'], 1)
data.head(10)
```

#### День недели, месяц и год публикации объявления.

```python
#найдем день недели публикации объявления
data['day_of_week_exposition'] = data['first_day_exposition'].apply(lambda x: x.strftime('%A'))
#найдем месяц публикации объявления
data['month_exposition'] = data['first_day_exposition'].apply(lambda x: x.strftime('%B'))
#найдем год публикации объявления
data['year_exposition'] = data['first_day_exposition'].dt.year
data.head(10)
```

#### Этаж квартиры; варианты — первый, последний, другой.

```python
def variation(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    elif row['floor'] > row['floors_total']:
        return 'ошибка'
    else:
        return 'другой'
data['floor_options'] = data.apply(variation, axis=1)
data.head(20)
```

#### Соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
#соотношение жилой к общей площади
data['living_area_corr'] = round(data['living_area'] / data['total_area'], 1)
#соотношение площади кухни к общей площади
data['kitchen_area_corr'] = round(data['kitchen_area'] / data['total_area'], 1)
data.head(20)
```
