#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Подготовка

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import GridSearchCV
from statsmodels.tsa.seasonal import seasonal_decompose
from itertools import product
warnings.filterwarnings('ignore')
```

```python
#преобразуем дату в тип datetime, сделаем её индексной и отсортиуем
data = pd.read_csv('/datasets/taxi.csv', parse_dates=[0], index_col=[0])
data.sort_index(inplace=True)
data.info()
data.head(5)
```

Данные в таблице записаны каждые 10 минут, т.е количество заказов измеряется в течение 10 минут. 

Период с 1 марта 2018 по 31 августа 2018.

## Анализ

```python
#ресемплирование по часам с суммированием заказов
data = data.resample('1H').sum()
figsize = (14,7)
data.plot(figsize=figsize)
```

```python
#посмотрим на тренд, сезонную составляющую и остатки

decomposed = seasonal_decompose(data, freq = ((24*60)//10))
plt.figure(figsize=(20, 10))

plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend', size=16)

plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality', size=16)

plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals', size=16)

plt.tight_layout()
```

```python
#проверим сезонность за неделю
decomposed.seasonal['2018-05-01':'2018-05-08'].plot(figsize= (14,7))
```

На графиках заметно постепенное увеличение количества заказов с небольшими колебаниями на протяжении всего периода.

По тренду видно, что всё больше людей начали использовать такси.

Количество заказов такси зависит от активности людей в течение суток. На протяжении недели видно, что к ночи количество заказов увеличивается, а утром наблюдается резкий спад.

Временной ряд в таком виде не является стационарным, так как подвержен тренду, который меняется во времени.

## Обучение

### Подготовка данных

```python
#создание дополнительных признаков для прогнозирования временного ряда

def make_features(data, rolling_mean_size):
    data = data.copy()
    
    data.columns = ["y"]

    data['lag_1']=data.y.shift(1)
    for i in range(24, 169, 24):
        data["lag_{}".format(i)] = data.y.shift(i)
       
    data['rolling_mean'] = data['y'].shift().rolling(rolling_mean_size).mean()
    data['hour'] = data.index.hour
    data['weekday'] = data.index.weekday
    data['is_weekend'] = data.weekday.isin([5,6])*1
    
    return data
```

```python
data_with_features = make_features(data, 24)
target = data_with_features.drop(columns = 'y')
features = data_with_features.y
```

```python
target_train, target_test, features_train, features_test = train_test_split(target, features, test_size = 0.1, shuffle = False)
target_train.dropna().index[0]
target_train = target_train.dropna()
features_train = features_train.loc['2018-03-08 00:00:00':]
```

### Выбор модели

В выборе лучшей модели представлены:

- DecisionTreeRegressor
- RandomForestRegressor
- LGBMRegressor
- CatBoostRegressor

```python
models = [DecisionTreeRegressor(random_state=12345),
          RandomForestRegressor(random_state=12345),
          LGBMRegressor(random_state=12345),
          CatBoostRegressor(random_state=12345)
         ]

param_grid = [{'max_depth': [4, 8, 16]},
              
              {'n_estimators': [50, 100, 200],
               'max_depth': [5, 10, 15]},
              
              {'max_depth': [2, 4, 8],
               'num_leaves': [2, 4, 8],
               'learning_rate': [0.1, 0.3, 0.5]},
              
              {'learning_rate': [0.1, 0.3, 0.5],
               'verbose': [False]},
              
              ]

cv = TimeSeriesSplit(n_splits=8)
results = []
best_score = 10000
best_model = None
for i, model in enumerate(models):
    grid_search = GridSearchCV(estimator=model,
                          param_grid=param_grid[i],
                          cv = cv,
                          scoring = 'neg_mean_squared_error')
    result = grid_search.fit(target_train, features_train)
    cv_results = pd.DataFrame(result.cv_results_)[['mean_test_score',
                                                   'std_test_score',
                                                   'mean_fit_time',
                                                   'mean_score_time',
                                                   'rank_test_score']]
    cv_results['Model'] = str(model).split('(')[0]
    cv_results['RMSE'] = np.sqrt(abs(cv_results['mean_test_score']))
    if cv_results['RMSE'].min() < best_score:
        best_score=cv_results['RMSE'].min()
        best_model = result.best_estimator_
    results.append(cv_results)
```

```python
all_models = pd.concat([result for result in results], axis=0).sort_values(by='RMSE', ascending = True).reset_index(drop=True)
all_models.head(10)
```

```python
print('Лучшая модель:', best_model)
```

## Тестирование

```python
predictions = best_model.predict(target_test)
print("RMSE на тестовой выборке", np.sqrt(mean_squared_error(features_test, predictions)))
```

**Вывод:**
По итогам сравнительной таблицы, лучшей моелью оказалась  RandomForestRegressor, которая имеет достаточно низкое RMSE равное 40,97. Эта модель в среднем ошибается на 40-41 заказа такси в час.
